```python
import os
import PyPDF2
import re
from langchain import OpenAI, LLMChain, PromptTemplate
# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    pdf_text = ""  
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            pdf_text += page.extract_text()
    return pdf_text

# Function to extract questions from the extracted text
def extract_questions(pdf_text):
    questions = re.findall(r'\d+\.\s+([^\n]+)', pdf_text)
    return questions

# Initialize the OpenAI model (set your API key here)
os.environ["OPENAI_API_KEY"] = ""  #enter your api key
llm = OpenAI(model_name="gpt-3.5-turbo")

# Define a prompt template for answering questions
template = PromptTemplate(input_variables=["question"], template="Q: {question}\nA:")

# Create an LLMChain with the template and language model
chain = LLMChain(prompt=template, llm=llm)

# Function to generate answers for the extracted questions
def generate_answers(questions):
    answers = []
    for question in questions:
        answer = chain.run({"question": question})
        answers.append(answer)
    return answers

# Main function to execute the entire process
def main(pdf_path):
    # Extract text from PDF
    pdf_text = extract_text_from_pdf(pdf_path)

    # Extract questions from the text
    questions = extract_questions(pdf_text)

    # Print extracted questions
    print("Extracted Questions:")
    for i, question in enumerate(questions, 1):
        print(f"{i}. {question}")
    # Generate answers for the questions
    answers = generate_answers(questions)

    # Print generated answers
    print("\nGenerated Answers:")
    for i, (question, answer) in enumerate(zip(questions, answers), 1):
        print(f"Q{i}: {question}")
        print(f"A{i}: {answer}\n")

# Execute the main function with the path to your PDF file
if __name__ == "__main__":
    pdf_path = "/content/Python MCQ Exam DataType 1.pdf"
    main(pdf_path)
```
